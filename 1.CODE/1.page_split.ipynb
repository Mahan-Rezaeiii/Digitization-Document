```python
import fitz
import os
```

```python
input_pdf = "/Users/mahan/Desktop/HC_Lab/Sanjesh-NoteBook/Sanjesh_Notebook/Digitization-PDF/Riazi/IN_PDF/sanjesh_riazi_91.pdf"
output_dir = "/Users/mahan/Desktop/HC_Lab/Sanjesh-NoteBook/Sanjesh_Notebook/Digitization-PDF/Riazi/OUT_PDF/"
```

```python
pdf = fitz.open(input_pdf)

# (output_filename, from_page, to_page)
split = [
    ("1-Roozaneh.pdf",        19, 40),
    ("2-Mahroum.pdf",         40, 41),
    ("3-Tarbiat_Moalem.pdf",  42, 45),
    ("4-Nimeh_Motemarkez.pdf",45, 47),
    ("5-Shabaneh.pdf",        48, 58),
    ("6-Nimeh_Hozouri.pdf",   59, 60),
    ("7-Payam_Nour.pdf",      61, 87),
    ("8-Gheir_Entefaii.pdf",  88,107),
    ("9-Majazi.pdf",          108,108),]

for filename, start, end in split:
    new_pdf = fitz.open()
    new_pdf.insert_pdf(pdf, from_page=start, to_page=end)
    
    outpath = os.path.join(output_dir, filename)
    new_pdf.save(outpath)
    new_pdf.close()

pdf.close()
```
